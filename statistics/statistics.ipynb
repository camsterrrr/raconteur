```python
tactic_file_name_mappings = {
    "collection_techniques": "Collection",
    "command_and_control_techniques": "Command and Control",
    "credential_access_techniques": "Credential Access",
    "defense_evasion_techniques": "Defense Evasion",
    "discovery_techniques": "Discovery",
    "execution_techniques": "Execution",
    "exfiltration_techniques": "Exfiltration",
    "impact_techniques": "Impact",
    "initial_access_techniques": "Initial Access",
    "lateral_movement_techniques": "Lateral Movement",
    "persistence_techniques": "Persistence",
    "privilege_escalation_techniques": "Privilege Escalation",
    "reconnaissance_techniques": "Reconnaissance",
    "resource_development_techniques": "Resource Development"
}

def fill_tactic(technique_: str, path: Path) -> str:
    ret_val = ""
    
    if (
        technique_ == "Other"
        or technique_ == "Unknown"
    ):
        return technique_
    
    for csv_file in os.listdir(path):
        working_path = path / Path(csv_file)
        tactic = working_path.stem    # Files are ordered by Tactic. This isolates the name of tactic.
        tactic = tactic_file_name_mappings[tactic]
        
        csv_df = pd.read_csv(working_path)
        
        for technique in csv_df["ID"]:
            if technique == technique_:
                ret_val = tactic
                break
    
    if ret_val == "":
        print(f"Edge case found: {technique_}")
    
    return ret_val 
    
```
